```python
import matplotlib.pyplot as plt
#%matplotlib inline
import numpy as np
```

```python
x=np.arange(1,5)
#plt.plot(x,x,'-r',linewidth=2,label="original")
#plt.plot(x,x*x,'-*g',linewidth=5,label="square")
plt.plot(x,np.log(x),'-vb',linewidth=1,markersize=5,label="log")
plt.xlabel("some numbers")
plt.ylabel("square cube")
plt.legend()
#plt.axis([0,6,0,26])
plt.show()
```

```python
fig, axes1 = plt.subplots(3,2,figsize=(30,20))
ar=np.arange(1,100)
axes1[0,0].plot(ar)
axes1[0,1].plot(ar,ar*ar,'r')
axes1[1,0].plot(ar,np.log(ar),'g')
theta=np.arange(0,2*np.pi+np.pi/3,np.pi/3)
axes1[1,1].plot(np.sin(theta),np.cos(theta),'k')
theta1=np.arange(0,2*np.pi+np.pi/2,np.pi/2)
axes1[2,0].plot(np.sin(theta1),np.cos(theta1),'navy')
year=[2014,2015,2016,2017,2018,2019,2020,2021,2022]
rel_price=[1000,1100,1250,1450,1600,1900,2100,2300,2500]
axes1[2,1].plot(year,rel_price)
```
