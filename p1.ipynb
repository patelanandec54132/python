```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
im=io.imread("cameraman.tif")
plt.figure(1)
plt.subplot(1,2,1)
plt.imshow(im,cmap='gray')
log = np.log(im+1)
logim=np.array(log,dtype=np.uint8)
plt.subplot(1,2,2)
plt.imshow(logim,cmap='gray')
```

```python
ilog=np.exp(im)-1
ilogim=np.array(ilog,dtype=np.uint8)
plt.figure(1)
plt.subplot(1,2,1)
plt.imshow(ilogim,cmap='gray')
```

```python
height,width=im.shape
ine=255-im  
fig = plt.figure()
fig.set_figheight(15)
fig.set_figwidth(15)
fig.add_subplot(1,2,1)
plt.imshow(im,cmap='gray')
plt.title("original")
fig.add_subplot(1,2,2)
plt.imshow(ine,cmap='gray')
plt.title("negetive")
plt.savefig('original-negetive.jpeg')    
```

```python
rev=im[::-1]
plt.subplot(1,2,1)
plt.imshow(im,cmap='gray')
plt.title("original")
plt.subplot(1,2,2)
plt.imshow(rev,cmap='gray')
plt.title('rotated')
plt.savefig('original-rotated.jpeg')    
```

```python
crop=im[100:200,150:250]
plt.subplot(1,2,1)
plt.imshow(im,cmap='gray')
plt.title('original')
plt.subplot(1,2,2)
plt.imshow(crop,cmap='gray')
plt.title('cropped')
plt.savefig("original-cropped")
```
